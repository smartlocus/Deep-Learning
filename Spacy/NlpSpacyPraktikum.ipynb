```python
# Maintainer Robiel Elias Redie 
import os 
import pandas as pd 
import spacy
from spacy import displacy
```

```python
# Laden der Daten
train_data = pd.read_csv('/home/robi/uni/KI/Praktikum6Spacy/QA_train_data.csv')
test_data = pd.read_csv('/home/robi/uni/KI/Praktikum6Spacy/QA_test_data.csv')
```

```python
nlp = spacy.load('en_core_web_md')
# loading the spacy english model 
```

```python
def predict_answer(question):
    # Process the question using spaCy
    question_doc = nlp(question)

    # Calculate similarity scores with training data question and imput question 
    similarity_scores = train_data['Question'].apply(lambda x: question_doc.similarity(nlp(x)))

    # Get the index of the most similar question
    related_question_index = similarity_scores.idxmax()

    return train_data['Answer'].iloc[related_question_index]
```

```python
# Predict answers for questions in the test data
test_data['Predicted_Answer'] = test_data['Question'].apply(predict_answer)
```

```python
display_table = test_data[['Question', 'Answer', 'Predicted_Answer']]
print(display_table)
```

```python
correct_predictions = sum(test_data['Predicted_Answer'] == test_data['Answer'])
total_questions = len(test_data)
```

```python
# Calculation of the Prediction Score
prediction_score = correct_predictions / total_questions
```

```python
# Display the Prediction Score
print(f'Prediction Score: {prediction_score}')
```
